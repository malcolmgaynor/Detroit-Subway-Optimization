```python
import geopandas as gpd

# Load the shapefile
file_path = "tl_2024_26_tract.shp"
census_data = gpd.read_file(file_path)

# View the first few rows
print(census_data.head())
```

```python
census_data['centroid_lat'] = census_data.geometry.centroid.y
census_data['centroid_lon'] = census_data.geometry.centroid.x

census_data
```

```python
detroit_df = detroit_df.merge(census_data[['GEOID','centroid_lat','centroid_lon']], left_on='GEOID', right_on='GEOID')
```

```python
detroit_df.Census_Tract.value_counts()
```

```python
aggregated_data = detroit_df.groupby('Census_Tract').agg(
    Total_C000=('C000', 'sum'),                # Sum of C000
    centroid_lat=('centroid_lat', 'first'),   # Retain the first centroid_lat
    centroid_lon=('centroid_lon', 'first')    # Retain the first centroid_lon
).reset_index()

aggregated_data.to_csv('work_data.csv', index=False)
```
