```julia
using CSV, DataFrames

df = CSV.read("assigned_data_with_distances.csv", DataFrame)
```

```julia
using StatsBase

# Standardizing columns
standardized_df = DataFrame(zscore.(eachcol(df)), names(df))

standardized_df
```

```julia
#Step 1: zero out data where they are already at the closest station (I ran it with and without this step, I don't think it matters

for i in 1:size(df, 1)
    col_name = Symbol("Num_worked_station_$i")
    df.Num_lived_all_new[i]=df.Num_lived_all_new[i]-df[i,col_name]
    df[i, col_name] = 0
end
df
```

```julia
# Function to calculate Euclidean distance
function euclidean_distance(coord1, coord2)
    sqrt((coord1[1] - coord2[1])^2 + (coord1[2] - coord2[2])^2)
end
```

```julia
# Define indices for stations
I = size(df)[1] 
J = size(df)[1] 
K = size(df)[1]
L = size(df)[1]
M = 10000 #large number, for constraint

p = df.Num_lived_all_new #p[i] = num ppl who live at station i 

b = df[:, end-(size(df)[1]-1):end] #b[h,w] = num ppl with home at station h who work as station w
distances = df[:,1:2]
distmatrix = DataFrame(fill(0.0, size(df)[1], size(df)[1]),:auto)
for first in 1:size(df)[1]
    for second in 1:size(df)[1]
        distmatrix[first,second] = euclidean_distance(distances[first,:],distances[second,:])
    end
end
distmatrix; # distmatrix[i,j] is the distance from i to j 
```

```julia
connectionlimit = 2
using Gurobi
using JuMP

# Initialize the model
model = Model(Gurobi.Optimizer)

# Decision variables
@variable(model, x[i=1:I,j=1:J], Bin) # Binary variable, 1 if connection is active between stations i and j
@variable(model, f[i=1:I,j=1:J,l=1:L,z=0:1] >=0 ) # Flow of people who live at station l between station i and j

### Question: does "double counting" matter??? I think it will inflate our obj fxn but I don't think it will actually matter in terms of the optimal value

# Objective function
# we want to minimize the sum of the flow between each station from people starting at each home location multiplied by the corresponding distance
@objective(model, Min, sum(sum( sum( sum( (f[i,j,l,z])*distmatrix[i,j] for i in 1:I) for j in 1:J) for l in 1:L) for z in 0:1))

#to ensure our indexing is consistent 
@constraint(model, [i=1:I, j=1:J], x[i,j]==x[j,i])

# Constraints
# the flow out from each starting population is equal to that starting population (even if the flow is circular for those working at home)
@constraint(model, [l=1:L], sum(f[l,j,l,1] for j in 1:J) == p[l] )

@constraint(model, [l=1:L,i=1:I], sum(f[i,j,l,0] for j in 1:J) >= b[l,i] ) ###must end up at work 


# if we don't build the connection, the flow must be zero 
@constraint(model, [i=1:I,j=1:J,z=0:1,l=1:L], f[i,j,l,z] <= M*x[i,j])

# Max of two outward connections for each line 
for j in 1:J
    @constraint(model, sum(x[i, j] for i in 1:I if i!=j) == connectionlimit)
end

for j in 1:J
    @constraint(model, sum(sum(f[i,j,l,0] for i in 1:I if i != j) for l in 1:L) == sum( sum( f[i,j,l,1] for i in 1:I if i !=j) for l in 1:L) + sum(b[l,j] for l in 1:L))
end




            
# Solve the model
optimize!(model)

# Output results
if termination_status(model) == MOI.OPTIMAL
    println("Optimal solution found!")
    println("Objective value: ", objective_value(model))
    println("Station decisions: ", value.(x))
else
    println("No optimal solution found.")
end
```
